### Preview: This notebook uses a [Kaggle Dataset](https://www.kaggle.com/tobycrabtree/nfl-scores-and-betting-data), a [538 dataset](https://github.com/fivethirtyeight/data/tree/master/nfl-elo) and a [reddit dataset](https://www.reddit.com/r/sportsbook/comments/3ep6q4/equation_to_convert_point_spread_to_oddsmoneyline/cthzcxd?utm_source=share&utm_medium=web2x) to see whether it is possible to predict which games to bet on depending on the stadium, ELO rating, and other factors. 

### Data Description - There are three main data sources:
- [Kaggle](https://www.kaggle.com/tobycrabtree/nfl-scores-and-betting-data) - This dataset contains 3 tables: ***stadiums, teams,*** and ***scores***. ***Stadiums*** has facts about every NFL stadium, including the type (indoor or outdoor), the weather type (i.e. dome) and the surface (i.e. turf). ***Teams*** contains each teams full name, division, conference, etc. ***Scores*** is the meat of this dataset - it contains every historical NFL game, including the score, the spread, the over-under, week #, stadium, and weather.
- [538 dataset](https://github.com/fivethirtyeight/data/tree/master/nfl-elo) - Contains 538's ratings for every NFL team pre- and post-game, qb elo rating pre- and post-game, and predictions for who will win each game

#### Add in the return on betting $100 on the favorite or the underdog

```python
def ml_to_profit(ml):
    if ml < 0:
        return 10000/ml*-1
    return ml
```

```python
scores["return_on_100_f"] = scores.apply(lambda x: ml_to_profit(x['moneyline_f']), axis=1)
scores["return_on_100_u"] = scores.apply(lambda x: ml_to_profit(x['moneyline_u']), axis=1)
```

```python
scores["team_winner"] = scores.apply(lambda x: x.team_home if x.home_win else x.team_away, axis=1)
```

The function below uses joined and teams_dict

```python
def graph_team(team, seasons=[2019], stat='elo', return_df=False):
    stat_dict = {'elo':['elo1_pre', 'elo2_pre'], 'prob':['elo_prob1', 'elo_prob2'], 
                 'spread':['spread_favorite', 'spread_underdog'], 'ml':['moneyline_f', 'moneyline_u'], 
                }
    extra = ['return on spread', 'return on ml', 'point difference']
    possible_stats = [stat for stat in stat_dict.keys()] + extra
    assert stat in possible_stats, "Make sure stat is in {}".format(possible_stats)
    assert team in teams_dict.values(), "Make sure to use the team ID instead of the team name."
    df = joined[((joined.team1 == team) | (joined.team2 == team)) & (joined.season.isin(seasons))]
    if stat in ['elo', 'prob']:
        df[stat] = df.apply(lambda x: x[stat_dict[stat][0]] if x.team1 == team 
                            else x[stat_dict[stat][1]], axis=1)
    elif stat in ['spread', 'ml']:
        df[stat] = df.apply(lambda x: x[stat_dict[stat][0]] if x.team_favorite_id == team 
                            else x[stat_dict[stat][1]], axis=1)
    elif stat == 'return on ml':
        df['return_on_100'] = df.apply(lambda x: return_col_ml(x, team), axis=1)
        df[stat] = df.return_on_100.cumsum()
        print("Final ML cumulative return on {} is {}".format(team, df[stat].iloc[-1]))
    elif stat == 'return on spread':
        df['return_on_100'] = df.apply(lambda x: return_col_spread(x, team), axis=1)
        df[stat] = df.return_on_100.cumsum()
        print("Final spread cumulative return on {} is {}".format(team, df[stat].iloc[-1]))
    else:
        df[stat] = df.apply(lambda x: x.score1 - x.score2 if x.team1 == team else x.score2 - x.score1 ,axis=1)
    df.plot(kind='line', x='date', y=stat)
    if return_df:
        return df
```

```python
def return_col_ml(df, team):
    if (df.team_favorite_id == team) and ((df.team1 == team and df.home_win == 1) or 
                                          (df.team2 == team and df.home_win == 0)):
        return df['return_on_100_f']
    elif (df.team_favorite_id != team) and ((df.team1 == team and df.home_win == 1) or 
                                          (df.team2 == team and df.home_win == 0)):
        return df['return_on_100_u']
    return -100
```

```python
def return_col_spread(df, team):
    if (df.team_favorite_id == team) and ((df.team1 == team and df.score1 - df.score2 > df.spread_favorite*-1) or 
                                          (df.team2 == team and df.score2 - df.score1 > df.spread_favorite*-1)):
        return 100
    elif (df.team_favorite_id != team) and ((df.team1 == team and df.score1 - df.score2 > df.spread_favorite) or 
                                          (df.team2 == team and df.score2 - df.score1 > df.spread_favorite)):
        return 100
    return -110
```
